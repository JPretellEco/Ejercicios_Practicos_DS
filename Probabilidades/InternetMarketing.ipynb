# Cul es la probabilidad de que visitante de FashionTrend visite la pgina de Dillard

```python

from sympy import symbols,Eq,solve

p = symbols('p') # Definimos la variable 

PMD = 3*p # La probabilidad de que el visitante de Dillar sea mujer es 3 veces ms que la de un hombre
PHD = p # Probabilidad de que el visitante de Dillar sea hombre

PD = PMD*PM + PHD*PH # Probabilidad de que se visite la pgina de Dillard

PD

print("La probabilidad de que un visiten la pgina de Dillard es: ",PD)
```

# Probabilidad de que el visitante de Dillar sea Mujer

### Teorema de Bayes Aplicado al Problema




```python
PDM = (PMD*PM)/PD # Probabilidad de que el visitante de Dillar sea mujer dado que se visit la pgina de Dillard
print("La probabilidad de que el visitante de Dillar sea mujer dado la probabilidad de que  visite la pgina de Dillard  \n es: ",round(PDM,4))

# Para redondear un float a 3 decimales se puede usar el mtodo que no sea round 
```
